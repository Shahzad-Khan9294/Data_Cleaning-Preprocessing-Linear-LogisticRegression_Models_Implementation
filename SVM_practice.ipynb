```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score

# Create a grid of values for the three features
x1_range = np.linspace(x_train[:, 0].min(), x_train[:, 0].max(), 50)
x2_range = np.linspace(x_train[:, 1].min(), x_train[:, 1].max(), 50)
x3_range = np.linspace(x_train[:, 2].min(), x_train[:, 2].max(), 50)

# Create a meshgrid for the three features
x1, x2, x3 = np.meshgrid(x1_range, x2_range, x3_range)

# Flatten the meshgrid
grid_points = np.c_[x1.ravel(), x2.ravel(), x3.ravel()]

# Compute the decision function for the grid points
decision_values = clf.decision_function(grid_points)

# Reshape decision function values to match the meshgrid
decision_values = decision_values.reshape(x1.shape)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot decision boundary
contour = ax.contourf(x1[:, :, 0], x2[:, :, 0], x3[:, :, 0], decision_values[:, :, 0], levels=[0], cmap='coolwarm', alpha=0.5)

# Plot training points
scatter = ax.scatter(x_train[:, 0], x_train[:, 1], x_train[:, 2], c=y_train, edgecolor='k')

ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')

plt.show()
```
